```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import fontstyle
# Read the data set into a Pandas dataframe
df = pd.read_csv('1982-2022.csv')

# Calculate the mean temperature for the entire data set

mean_temp = df['Value'].mean()

# Calculate the change in temperature between successive time periods
temp_diff = df['Value'].diff()

# Calculate the sum of the temperature change values
temp_increase = temp_diff.sum()

all_years = df["Year"]
values = df["Value"]

plt.bar(all_years, values)
plt.title('Change in Temperature Over Time')
plt.xlabel('Years')
plt.ylabel('Temperature Change (Â°C)')
plt.xticks(rotation=45)
plt.show()
text ='OVERALL ICREASE IN TEMPERATURE:'

text1=fontstyle.apply(text,'bold/Serif/BLUE_BG')
text2=fontstyle.apply(temp_increase,'bold/Serif/BLUE_BG')
print (text1 + text2)
```
