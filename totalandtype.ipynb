```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data_1=pd.read_excel("C:/Users/umutg/Desktop/gaz.xls")
```

```python
data_1=data_1.dropna()
```

```python
data_1.columns
```

```python
df_1=data_1.rename(columns={"Unnamed: 0":"Year","Unnamed: 1":"Total","Unnamed: 2":"CO2","Unnamed: 3":"CH4","Unnamed: 4":"N2O","Unnamed: 5":"F-gases"})
```

```python
df_1=df_1.drop(labels=2,axis=0)
```

```python
df_1['Year']=df_1['Year'].astype('int')
```

```python
plt.figure(figsize=(10, 6))
plt.title("Greenhouse gas emissions in Turkey")
plt.xlabel("Year")
plt.ylabel("Emissions (million metric tons of CO2 equivalent)")

plt.plot(df_1["Year"], df_1["Total"], label="Total")
plt.plot(df_1["Year"], df_1["CO2"], label="CO2")
plt.plot(df_1["Year"], df_1["CH4"], label="CH4")
plt.plot(df_1["Year"], df_1["N2O"], label="N2O")
plt.plot(df_1["Year"], df_1["F-gases"], label="F-gases")
plt.legend()
plt.show()
```

```python
df_2=df_1.copy()
df_2.set_index("Year")
for col in df_2.columns:
    if col=='Year':
        continue
    plt.figure(figsize=(15, 5))
    plt.plot(df_2["Year"], df_2[col])
    plt.title(col)
    plt.show()
```
